```python
def extraiDadosOpcao(opcao):  # Apenda cotacao e vencimento de cada opcao
    #print('Extraindo dados de:',opcao)
    import time
    time.sleep(sleep)
    index = Opções.index(opcao)
    try:
        response = requests.get('https://opcoes.net.br/' + opcao)
        pagina = BeautifulSoup(response.text, 'html.parser')

        # Cotação = []
        cotacao = pagina.find_all('table', attrs={'class': 'table table-bordered table-condensed top-buffer-20'})[0].find_all('tr')[2].find_all('td')[4].text
        cotacao = float(cotacao.replace(',', '.'))
    except:
        cotacao = float(0)
    Cotação[index] = cotacao

    # investimentoMin = []
    investMin = float(cotacao*100)
    investimentoMin[index] = investMin

    try:
        # DataVencimento = []
        vencimento = pagina.find_all(
            'div', attrs={'class': 'col-md-12'})[0].find_all('h1')[0].text
        vencimento = vencimento[-10:]
        DataVencimento[index] = vencimento
    except:
        vencimento = '31/12/9999'

    try:
        # volFinanceiro = []
        volFin = pagina.find_all('table', attrs={
                                    'class': 'table table-bordered table-condensed top-buffer-20'})[0].find_all('tr')[2].find_all('td')[6].text
        volFin = volFin.replace('.', '').replace(',', '.')
    except:
        volFin = 0
    volFinanceiro[index] = float(volFin)
    try:
        volatImpl = pagina.find_all('table', attrs={'class' : 'table table-bordered table-condensed top-buffer-20'})[0].find_all('tr')[2].find_all('td')[11].text.replace('.', '').replace(',', '.')
        volatImplicita[index] = float(volatImpl)
    except:
        volatImplicita[index] = 0
```

```python
def diferencaEmDias(opcao):#Retorna lista de diferença entre duas datas em dias
    #print(opcao)
    from networkdays import networkdays
    import datetime    
    index = Opções.index(opcao)
    try:
        delta = networkdays.Networkdays(datetime.date(datetime.datetime.now().year,
                                        datetime.datetime.now().month,
                                        datetime.datetime.now().day),
                                        datetime.date(int(DataVencimento[index][6:10]),
                                        int(DataVencimento[index][3:5]),
                                        int(DataVencimento[index][0:2]))
        )
        diferencaDatas[index] = len(delta.networkdays())
        #print('A opção',opcao,'vence em',diferencaDatas[index],'dias')
    except:
        diferencaDatas[index] = 0
```

```python
def calculaPerformance(opcao):
    try:
        #time.sleep(0.5)
        index = Opções.index(opcao)
        cotacaoAtual = PrecoAtual[index]
        acao = Ticker[index]
        tipo = Tipo[index]
        tabelaOpcoes = pd.DataFrame(Análise)

        taxa = round(abs(math.pow((Strike[index]/cotacaoAtual),(1/diferencaDatas[index]))-1)*100,2)
        TaxaDiaria[index] = taxa

        distancia = round((Strike[index]/PrecoAtual[index]-1)*100,2)
        distPreco[index] = distancia
        
        data = DataVencimento[index]
        cotacaoMax = tabelaOpcoes['Cotação'].loc[tabelaOpcoes['Vencimento'] == data].loc[tabelaOpcoes['Opção'].str.contains(acao[0:4],case=True)].loc[tabelaOpcoes['Tipo'] == tipo].max()
        potencial = round(abs(cotacaoMax/Cotação[index]),2)
        potencialValorizacao[index] = potencial

        performance = potencialValorizacao[index]/(TaxaDiaria[index]*abs(distPreco[index]))+volFinanceiro[index]
        Performance[index] = performance

        #print('Performance da',opcao,'calculada com sucesso')
    except:
        Performance[index] = 0
        #print('Erro ao calcular a performance da',opcao)
```

```python
def main():
    for acao in listaAcoes:
        extraiSiglasPut(acao)
        extraiSiglasCall(acao)
    for opcao in Opções:
        extraiDadosOpcao(opcao)
        diferencaEmDias(opcao)
        calculaPerformance(opcao)
```
